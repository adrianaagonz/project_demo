## 8. Análisis univariado de variables categóricas
Realizar gráficos de barras para analizar la frecuencia de las variables categóricas.

```python
# Bar plot for a categorical variable
import matplotlib.pyplot as plt
import seaborn as sns 

# Preparamos los datos de las 10 mejores marcas:
top_brands = df['brand_name'].value_counts().head(10).reset_index()

# Dibujamos un gráfico de barras a todo color:
sns.barplot(data=top_brands, x='count', y='brand_name')
plt.title('Top 10 Brands with more products in Sephora')
plt.show()
```

## 9. Análisis bivariado entre variables
Explorar relaciones entre variables numéricas y categóricas mediante scatterplots y tablas cruzadas.

```python
# Scatterplot
# plt.scatter(df['numeric_column_x'], df['numeric_column_y'])
# plt.show()

# Crosstab for two categorical variables   
# pd.crosstab(df['categorical_column1'], df['categorical_column2'])
import matplotlib.pyplot as plt
import seaborn as sns 

# Dibujamos el gráfico cruzando la nota (rating) y los "me gusta" (loves_count)
sns.scatterplot(data=df, x='rating', y='loves_count', alpha=0.5, color='purple')
plt.title('Relation between "rating" and "loves" in Sephora') 
plt.xlabel('Nota (Rating)')
plt.ylabel('Cantidad de "Loves"')
plt.show()
```

## 10. Visualización de correlaciones
Calcular y visualizar la matriz de correlación entre variables numéricas usando un heatmap.

```python
import matplotlib.pyplot as plt
import seaborn as sns 

# Calculamos la correlación 
corr = df.corr(numeric_only=True)

# Dibujamos el mapa de calor )
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variable')
plt.show()


# Correlation matrix
# corr = df.corr()
# sns.heatmap(corr, annot=True)
# plt.show()
```
